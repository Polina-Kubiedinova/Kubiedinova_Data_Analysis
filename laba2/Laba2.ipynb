#### 1.1 Створення

```python
import numpy as np
scalar = 40
print("Скаляра:", scalar)
vector = np.arange(10)
print("Вектор:", vector)
matrix = np.random.randint(0, 10, (3,3))
print("Матриця:\n", matrix)
print(matrix.shape)
```

```python
print(matrix.shape)
print("\nЧитання елементів:")
print("Скаляра:", scalar)
print("Другий елемент вектора:", vector[1])
print("Елемент у другому рядку, третьому стовпці матриці:", matrix[2, 2])
```

#### 1.3 зміна

```python
vector[0] = 100 
matrix[2, 1] = 8 
print("\n Після оновлення:")
print("Вектор:", vector)
print("Матриця:\n", matrix)
```

#### 1.4 видалення

```python
vector = np.delete(vector, 3)  
matrix = np.delete(matrix, 0, axis=0) 
print("\nПісля видалення:")
print("Вектор:", vector)
print("Матриця:\n", matrix)
```

### 3.1 генерація даних

```python
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 20)
y_true = 2 * x + 3  
np.random.seed(42)
noise = np.random.normal(0, 2, size=x.shape)
y_measured = y_true + noise
```

### Будуємо графік

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y_true, label="Істинна функція", color="purple")
plt.scatter(x, y_measured, label="Виміряні дані", color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Лінійна функція з похибками вимірювань")
plt.grid(True)
plt.show()
```

### похибки

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y_true, y_measured)
mse = mean_squared_error(y_true, y_measured)

print("MAE:", mae)
print("MSE:", mse)
```

### результати

```python
results = pd.DataFrame({
   "Метрика": ["MAE", "MSE"],
   "Значення": [mae, mse]
})
results.to_csv("results.csv", index=False, encoding="utf-8-sig")
```
