#### Здійснити загальні операції з використанням TensorFlow :

```python
import tensorflow as tf

scalar = tf.constant(5)                       
vector = tf.constant([1, 2, 3, 4])             
matrix = tf.constant([[1, 2], [3, 4]])         
tensor = tf.constant([[[1, 2], [3, 4]], [[5,6],[7,8]]])

print("Vector shape:", vector.shape, "rank:", tf.rank(vector).numpy(), "size:", tf.size(vector).numpy())
print("Matrix shape:", matrix.shape, "rank:", tf.rank(matrix).numpy(), "size:", tf.size(matrix).numpy())
print("Tensor shape:", tensor.shape, "rank:", tf.rank(tensor).numpy(), "size:", tf.size(tensor).numpy())

t1 = tf.random.uniform([5, 300], minval=0, maxval=1)
t2 = tf.random.uniform([5, 300], minval=0, maxval=1)
print("\nt1.shape, t2.shape:", t1.shape, t2.shape)

matmul_result = tf.matmul(t1, t2, transpose_b=True)
print("\nMatrix multiplication result shape:", matmul_result.shape)

dot_result = tf.reduce_sum(t1 * t2)
print("\nDot product result:", dot_result.numpy())

image_tensor = tf.random.uniform([224, 224, 3], minval=0, maxval=1)
print("\nimage_tensor.shape:", image_tensor.shape)

print("\nMin value:", tf.reduce_min(image_tensor).numpy())
print("Max value:", tf.reduce_max(image_tensor).numpy())

batch_tensor = tf.random.uniform([1, 224, 224, 3])
print("\nTensor shape:", batch_tensor.shape)
squeezed_tensor = tf.squeeze(batch_tensor)
print("\nSqueezed tensor shape:", squeezed_tensor.shape)

tensor_10 = tf.constant([5, 2, 9, 1, 7, 6, 3, 8, 0, 4])
max_index = tf.argmax(tensor_10)
print("\nTensor:", tensor_10.numpy())
print("Index of max value:", max_index.numpy())

one_hot = tf.one_hot(max_index, depth=10)
print(one_hot.numpy())

```

#### Згенерувати набір даних для регресії та класифікації з використанням TensorFlow.

```python
x_reg = tf.random.uniform([200, 1], minval=0, maxval=10)
y_reg = 3 * x_reg + 7 + tf.random.normal([200, 1], mean=0.0, stddev=2.0)

from sklearn.datasets import make_classification

X_class0 = tf.random.normal([50, 2], mean=-2, stddev=1)
X_class1 = tf.random.normal([50, 2], mean=2, stddev=1)
X_class = tf.concat([X_class0, X_class1], axis=0)
y_class = tf.concat([tf.zeros(50), tf.ones(50)], axis=0)
```

#### Здійснити графічний аналіз наборів даних за допомогою Matplotlib.

```python
plt.figure(figsize=(8,5))
plt.scatter(x_reg.numpy(), y_reg.numpy(), color='blue', alpha=0.6)
plt.title("Регресійний набір даних")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

plt.scatter(X_class[:, 0], X_class[:, 1], c=y_class, cmap="coolwarm")
plt.title("Набір даних для класифікації")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```
